```python
tokenizer=AutoTokenizer.from_pretrained('meta-llama/Meta-Llama-3.1-8B-Instruct', true_remote_code=True)
```

```python
messages=[
    {'role':'system', 'content':'You are a helpful assistant.'},
    {'role':'user', 'content':'Tell me a lighth-hearted joke for a room of Data Scientist'}
  ]

prompt=tokenizer.apply_chat_template(messages,tokenize=False, add_generation_prompt=True)
print(prompt)
```

```python
PHI3_MODEL_NAME='microsoft/Phi-3-mini-4k-instruct'
QWEN2_MODEL_NAME='Qwen/Qwen2-7B-Instruct'
STARCOADER2_MODEL_NAME='bigcode/starcoder2-3b'
```

```python
phi3_tokenizer=AutoTokenizer.from_pretrained(PHI3_MODEL_NAME)
text='I am excited to show Tokenizers in actions to my LLM engineerings'
print()
print(phi3_tokenizer.encode(text))
print()
print(phi3_tokenizer.decode(phi3_tokenizer.encode(text)))
```

```python
print(tokenizer.apply_chat_template(messages, tokenize=False, add_generation_prompt=True))
print()
print(phi3_tokenizer.apply_chat_template(messages, tokenize=False, add_generation_prompt=True))
```

```python
qwen2_tokenizer=AutoTokenizer.from_pretrained(QWEN2_MODEL_NAME)
print(qwen2_tokenizer.encode(text))
print()
print(phi3_tokenizer.encode(text))
print()
print(tokenizer.encode(text))
```

```python
print(tokenizer.apply_chat_template(messages, tokenize=False, add_generation_prompt=True))
print()
print(phi3_tokenizer.apply_chat_template(messages, tokenize=False, add_generation_prompt=True))
print()
print(qwen2_tokenizer.apply_chat_template(messages, tokenize=False, add_generation_prompt=True))
```

```python
startcoader2_tokenizer=AutoTokenizer.from_pretrained(STARCOADER2_MODEL_NAME)
code = """
def hello_world():
  print("Hello World")
"""
tokens = startcoader2_tokenizer.encode(code)
for token in tokens:
  print(startcoader2_tokenizer.decode(token))
```
